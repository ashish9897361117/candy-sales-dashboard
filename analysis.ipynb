Data Cleaning


```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True) 
```

Check Null Values 


```python
df.isnull().sum()
```

```python
df.dropna()
```

Invalid rows remove

```python
df = df[df['Sales'] > 0]
```

```python
df = df[df['Units'] > 0]
```

Gross Margin %


```python
df['Gross Margin %'] = (df['Gross Profit'] / df['Sales']) * 100
```

Profit Per Unit

```python
df['Profit per Unit'] = df['Gross Profit'] / df['Units']
```

Revenue Contribution

```python
df['Revenue Contribution'] = df['Sales'] / df['Sales'].sum()
```

Cost vs Sales Scatter Plot

```python
import matplotlib.pyplot as plt
# it Shows the Margin Risk
plt.scatter(df['Cost'], df['Sales'])
plt.xlabel("Cost")
plt.ylabel("Sales")
plt.show()
```

High sales but low profit products?

```python
df["Profit Margin"] = df["Gross Profit"] / df["Sales"]
```

```python
df.groupby("Product Name")[["Sales","Gross Profit"]].sum().sort_values("Sales", ascending=False)
```

```python
result = df.groupby("Product Name")[["Sales","Gross Profit"]].sum()
result[(result["Sales"]>result["Sales"].median()) &
       (result["Gross Profit"]<result["Gross Profit"].median())]
```
